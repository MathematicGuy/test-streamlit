```python
import cv2
from matplotlib import pyplot as plt

hand_img = cv2.imread("hand.png",0)

plt.imshow(hand_img, cmap='gray')
plt.title("Hand_img")
plt.axis('off') 
plt.show()
```

```python
counts = dict()

height, width = hand_img.shape
for row in range(height):
    for col in range(width):
        counts[hand_img[row,col]] = counts.get(hand_img[row,col],0) + 1
```

```python
names = list(counts.keys())
values = list(counts.values())
plt.bar(range(len(counts)), values, tick_label=names)
plt.show()
```
